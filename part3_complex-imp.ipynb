```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
R = 2000
L = 5e-3
C = 300e-9
f = np.linspace(1, 20000, 20000)
z = 1/(1/R + 1/(2j*np.pi*f*L) + 2j*np.pi*f*C)

def g(f, R, L, C):
    return z
print(z)
```

```python
complexZ = g(f, R, L, C)
magZ = np.abs(complexZ)
argZ = np.angle(complexZ)
doubRcomplexZ = 1/(1/(R*2) + 1/(2j*np.pi*f*L) + 2j*np.pi*f*C)

doubRmagZ = np.abs(doubRcomplexZ)
doubRargZ = np.angle(doubRcomplexZ)

plt.figure(1)
plt.plot(f, magZ, 'r', label='2000Ω')
plt.plot(f, doubRmagZ, 'b', label='4000Ω')
plt.title('Magnitude')
plt.legend(loc=1)
plt.savefig('phys133_complex_mag.pdf')

plt.figure(2)
plt.plot(f, argZ, 'r', label='2000Ω')
plt.plot(f, doubRargZ, 'b', label='4000Ω')
plt.title('Phase')
plt.legend(loc=1)
plt.savefig('phys133_complex_arg.pdf')
```
